```python
import pandas as pd 
```

```python
##reading csv
month = pd.read_csv("BTCUSDT-1m-2022-07.csv")
day1 = pd.read_csv("BTCUSDT-1m-2022-07-24.csv")
day2 = pd.read_csv("BTCUSDT-1m-2022-07-25.csv")
day3 = pd.read_csv("BTCUSDT-1m-2022-07-27.csv")
day4 = pd.read_csv("BTCUSDT-1m-2022-07-28.csv")
day5 = pd.read_csv("BTCUSDT-1m-2022-07-30.csv")
day6 = pd.read_csv("BTCUSDT-1m-2022-07-31.csv")
```

```python
##naming as one variable
all = [day1,day2,day3,day4,day5,day6,month]
```

```python
##concatting
new = pd.concat(all)
```

```python
##setting as datetime and as index
new["open_time"] = pd.to_datetime(new["open_time"])
new = new.set_index("open_time")
```

```python
##sorting ascending
new = new.sort_values(by='open_time')
```

```python
new.to_csv("BTCUSDT-1m-2022-07.csv", index = False)
```
