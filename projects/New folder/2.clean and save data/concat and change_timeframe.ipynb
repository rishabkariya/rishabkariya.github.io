```python
import pandas as pd
import os
```

```python
# Set the directory path where your CSV files are located
directory_path = r'F:\3.price data\FUTURES\BTCUSDT\excel'

# Get a list of all CSV files in the directory
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

# Initialize an empty list to store dataframes
dataframes = []

# Loop through each CSV file and read it into a dataframe, then append to the list
for csv_file in csv_files:
    csv_path = os.path.join(directory_path, csv_file)
    df = pd.read_csv(csv_path)
    dataframes.append(df)

# Concatenate all dataframes into a single dataframe
df = pd.concat(dataframes, ignore_index=True)

print("CSV files concatenated'")
```

```python
# Load your CSV file with the modified open_time and initial 6 columns
#df = pd.read_csv("concated.csv")

# Convert the 'open_time' column to a datetime index
df['open_time'] = pd.to_datetime(df['open_time'])
df.set_index('open_time', inplace=True)

# Resample the data to 5-minute intervals and aggregate using OHLCV (Open, High, Low, Close, Volume)
df = df.resample('5T').agg({'open': 'first', 
                                      'high': 'max', 
                                      'low': 'min', 
                                      'close': 'last', 
                                      'volume': 'sum'})
```
