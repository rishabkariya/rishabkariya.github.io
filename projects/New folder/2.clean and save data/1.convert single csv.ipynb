```python
import pandas as pd
```

```python
name_csv = "BTCUSDT-1m-2020-01.csv
```

```python
##adds name to each column
df = pd.read_csv(name_csv,names =["open_time","open","high","low",
                                                  "close","volume","close_time",
                                                  "quote_volume","count","taker_buy_volume",
                                                  "taker_buy_quote_volume","ignore"])

##change time from unix to date time format
df["open_time"]=pd.to_datetime(df["open_time"],unit= "ms")

##select only initial 6 columns and saves in variable df
df = df.iloc[:, :6]
```

```python
df.to_csv(name_csv, index = False)
```
