```python
import pandas as pd
```

```python
# Load your CSV file with the modified open_time and initial 6 columns
df = pd.read_csv("BTCUSDT-1m-2020-02.csv")

# Convert the 'open_time' column to a datetime index
df['open_time'] = pd.to_datetime(df['open_time'])
df.set_index('open_time', inplace=True)

# Resample the data to 5-minute intervals and aggregate using OHLCV (Open, High, Low, Close, Volume)
df = df.resample('2T').agg({'open': 'first', 
                                      'high': 'max', 
                                      'low': 'min', 
                                      'close': 'last', 
                                      'volume': 'sum'})
```
