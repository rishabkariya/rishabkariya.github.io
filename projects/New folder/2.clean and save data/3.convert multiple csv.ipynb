```python
import pandas as pd
import os
```

```python
# Define the directory containing your CSV files
directory = r"F:\3.price data\FUTURES\BTCUSDT\excel"

# Get a list of CSV file names in the directory
csv_files = [filename for filename in os.listdir(directory) if filename.endswith(".csv")]

# Loop through the list of CSV files
for csv_file in csv_files:
    # Construct the full path to the CSV file
    file_path = os.path.join(directory, csv_file)
    
    try:
        # Read the CSV file and skip the header row
        df = pd.read_csv(file_path, skiprows=1, header=None)
        
        # Assuming 'open_time' is the first column
        df.rename(columns={0: "open_time", 1: "open", 2: "high", 3: "low", 4: "close", 5: "volume"}, inplace=True)
        
        # Change time from unix to datetime format
        df["open_time"] = pd.to_datetime(df["open_time"], unit="ms")
        
        # Select only the initial 6 columns
        df = df[["open_time", "open", "high", "low", "close", "volume"]]
        
        # Now you can perform any desired operations on this DataFrame
        # For example, you can save the modified DataFrame to a new CSV file
        output_file = f"modified_{csv_file}"
        df.to_csv(output_file, index=False)
        
        print(f"Processed {csv_file} and saved as {output_file}")
    except pd.errors.ParserError:
        print(f"Skipped {csv_file} due to existing columns")

print("All files processed successfully!")
```
